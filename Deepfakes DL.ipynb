```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.utils import to_categorical
import json

```

```python
metadata_file = "F:\\DL\\train_sample_videos\\metadata.json"

with open(metadata_file, 'r') as f:
    metadata = json.load(f)

# Load and preprocess video frames
real_frames = []  # List to store frames from real videos
fake_frames = []  # List to store frames from fake videos
```

```python
# Iterate through video files and load frames
for filename, metadata_entry in metadata.items():
    # Load frames, preprocess, and append to the appropriate list (real or fake)
    video = cv2.VideoCapture(os.path.join("F:\\DL\\train_sample_videos", filename))
    frames = []
    while True:
        ret, frame = video.read()
        if not ret:
            break
        frame = cv2.resize(frame, (128, 128))  # Resize frames as needed
        frames.append(frame)
    video.release()
    
    if metadata_entry['label'] == 'REAL':
        real_frames.extend(frames)
    elif metadata_entry['label'] == 'FAKE':
        fake_frames.extend(frames)
```

```python
# Create labels
real_labels = ['real'] * len(real_frames)
fake_labels = ['fake'] * len(fake_frames)

# Combine frames and labels
X = real_frames + fake_frames
y = real_labels + fake_labels
```

```python
# Encode labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_categorical = to_categorical(y_encoded)
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.2, random_state=42)
```

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(2, activation='softmax'))
```

```python
from keras.callbacks import EarlyStopping

# Define the EarlyStopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model with early stopping
model.fit(np.array(X_train), np.array(y_train), epochs=10, batch_size=16, validation_split=0.2, callbacks=[early_stopping])
```

```python
from tensorflow import keras

# Save your model in the native Keras format
model.save("your_model.keras")
```
